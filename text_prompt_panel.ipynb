```python
# Utils widgets
from ipywidgets import Button, Layout, jslink, IntText, IntSlider
from ipywidgets import AppLayout, Button, Layout
import ipywidgets as widgets
from ipywidgets import interactive
import asyncio
import nest_asyncio
nest_asyncio.apply()
from grpclib.client import Channel
from proto.seq2seq_grpc import Seq2SeqStub

from proto.seq2seq_pb2 import PromptRequest, PromptReply

def create_expanded_html(description, button_style):
    return widgets.HTML(description=description, button_style=button_style, layout=Layout(height='auto', width='auto'))

async def stream_prompt() -> None:
    channel = Channel(port = 50050)
    stub = Seq2SeqStub(channel)
    res = prompt.value
    # This block performs the same UNARY_STREAM interaction as above
    # while showing more advanced stream control features.
    async with stub.PromptModel.open() as stream:
        await stream.send_message(PromptRequest(prompt=res), end=True)
        #replies = [reply async for reply in stream]
        replies = list()
        message = ""
        async for reply in stream:
            # print(reply)
            message += reply.message
            replies.append(reply)
            value.value = message
        # print(replies)

    channel.close()

def _stream_prompt(b):
    loop = asyncio.get_event_loop()
    result = loop.run_until_complete(stream_prompt())


async def stream_redis_index() -> None:
    channel = Channel(port = 50050)
    stub = Seq2SeqStub(channel)
    async with stub.PromptModel.open() as stream:
        await stream.send_message(PromptRequest(prompt=prompt.value), end=True)
        #replies = [reply async for reply in stream]
        replies = list()
        message = ""
        async for reply in stream:
            #print(reply)
            message += reply.message
            replies.append(reply)
            #value.value = message
        #print(replies)

    channel.close()

def _stream_redis_index(b):
    loop = asyncio.get_event_loop()
    result = loop.run_until_complete(stream_prompt())


header_button = create_expanded_html('My Chat Bot', 'success')
footer_button = create_expanded_html('Footer', 'success')
options = ['open-ai-seq2seq', 'chat']
zips_dropdown  = widgets.Dropdown(
    options=options,
    value=options[0],
    description='Pipeline',
    disabled=False,
    )
prompt = widgets.Text(
    value=None,
    description='Prompt',
    disabled=False,
)
button = widgets.Button(
    description='Prompt',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Run report',
    icon='check' # (FontAwesome names without the `fa-` prefix)
)
button.on_click(_stream_prompt, False)
promp_template = widgets.HTML(
    description="Template",
    value="What is a good name for a company that makes {product}? List 5",
)
db_input = widgets.HTML(
    value="Input",
)
value = widgets.HTML(
    value="Ask question",
)
value.add_class('pre-wrap')


name = widgets.Text(description='Name:', padding=4)
color = widgets.Dropdown(description='Color:', padding=4, options=['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet'])
page1 = widgets.VBox(children=[zips_dropdown, promp_template, prompt, button, value], padding=4)

page2 = widgets.Box(children=[], padding=4)

tabs = widgets.Tab(children=[page1, page2])

tabs.set_title(0, 'Prompt')
tabs.set_title(1, 'Database')

AppLayout(header=header_button,
          left_sidebar=None,
          center=tabs,
          right_sidebar=None,
          footer=footer_button)
```
